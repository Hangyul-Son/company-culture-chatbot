```python
# prompt: upload a file to google drive and extract

import fitz  # PyMuPDF
import os

def extract_sections_from_pdf(file_path):
    try:
        doc = fitz.open(file_path)
        sections = []
        for page_num in range(doc.page_count):
            page = doc.load_page(page_num)
            text = page.get_text("blocks")  # Extract text blocks
            for block in text:
                sections.append(block)  # Append each block to the sections list
        return sections
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None
```

```python
extracted_sections = extract_sections_from_pdf("[카카오] 지속가능경영보고서 등 관련사항(자율공시) (2024.06.07).pdf")

if extracted_sections:
    # Process the extracted sections
    for i, section in enumerate(extracted_sections):
        # Print the content of each section
        print(section)
        # You can access other block information like its bounding box:
    # print(f"Bounding box: {section[0]}")
```

# Analyze the extracted sections using Gemini


```python
import google.generativeai as genai     
import os
import fitz
import json
```

```python
import dotenv

dotenv.load_dotenv()
```

```python
genai.configure(api_key=os.getenv("GOOGLE_API_KEY"))
model = genai.GenerativeModel("gemini-1.5-pro-002")
```

```python
# Extract text from PDF
doc = fitz.open("카카오 ESG 보고서 '2023 카카오의 약속과 책임'.pdf")
full_text = ""
for page_num in range(doc.page_count):
    page = doc.load_page(page_num)
    full_text += page.get_text()
```

```python

# Create prompt for GPT analysis
prompt = """Analyze the following text from a corporate report and extract:
1. Sections related to company's internal culture - compensation, benefits, work environment, team strcture, genders, age, average working years, etc. Even little bit related, extract them all.
2. Their headings
3. The corresponding content
4. As many sections as possible
5. Longer content the better, the content should be a paragraph or more.

Format the response as JSON with the following structure:
{
    "culture_sections": [
        {
            "heading": "section heading",
            "content": "section content"
        },
        ...
        {
            "heading": "section heading",
            "content": "section content"
        }
    ]
}

Text to analyze:
"""
```

```python
print(prompt + full_text)
```

```python
response = model.generate_content(prompt + full_text)
print(response.text)
```
